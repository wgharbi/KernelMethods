```python
def accuracy(pred, real):
    tot = 0.
    for i, val in enumerate(real):
        if val == pred[i]:
            tot += 1
    return tot/len(pred)
```

```python
def min_kernel(a, b):
    return np.sum(np.minimum(a,b))

def symmetrize(a):
    return a + a.T - np.diag(a.diagonal())

def compute_min_kernel(X):
    X = np.array(X)
    length = X.shape[0]
    ker = np.zeros((length, length))
    for i in range(length):
        for j in range(length):
            if i>=j:
                ker[i, j] = min_kernel(X[i], X[j])
    return ker     

def compute_test_ker(X_train, X_test):
    X_train = np.array(X_train)
    X_test = np.array(X_test)
    len1 = X_train.shape[0]
    len2 = X_test.shape[0]
    ker = np.zeros((len1, len2))
    for i in range(len1):
        for j in range(len2):
            ker[i,j] = min_kernel(X_train[i], X_test[j])
    return ker
```
